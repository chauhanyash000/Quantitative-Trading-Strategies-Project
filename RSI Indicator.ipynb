```python
raw=pd.read_csv('file:///Users/yashchauhan/Desktop/Quant Trading Project/%5ENSEI.csv', parse_dates=True)
data= (pd.DataFrame(raw[['Date','Close','Open']]).dropna())
Period=13;
n=data['Close'].count();
data.reset_index(drop=True, inplace=True)
```

```python
i=1;
while i<n:
    data.loc[i,'Change']=data.loc[i,'Close']-data.loc[i-1,'Close'];
    i+=1;
data['profit']=np.where(data['Change']>0,data['Change'],0)
data['Loss']=np.where(data['Change']<0,-data['Change'],0)
data['average_profit']=data['profit'].rolling(Period).mean()
data['average_loss']=data['Loss'].rolling(Period).mean()
data['rs']=data['average_profit']/data['average_loss']
data['rsi']=100-100/(data['rs']+1)
```

```python
data
```

```python
Data=pd.DataFrame(data[['Date','Open','Close','rsi']])
Data.dropna(inplace=True)
Data['Position']=0
Data.reset_index(drop=True, inplace=True)
n=Data['rsi'].count();
i=0;
while i<n:
    if Data.loc[i,'rsi']>70 and i<n:
        Data.loc[i,'Position']=-1;
        i+=1;
        while i<n:
            if Data.loc[i,'rsi']<30 and i<n:
                i+=1;
                break;
            Data.loc[i,'Position']=-1;
            i+=1;
    i+=1;
i=0;
while i<n:
    if Data.loc[i,'Position']==0:
        Data.loc[i,'Position']=1;
    i+=1;
```
